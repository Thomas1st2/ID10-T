```python
## Loop through the folders to get the files and append the data to an empty list before concatting the data to the DataFrame
## From here we drop any duplicates as the loop will concat the already stored data in the list.

dfs = []

for num in range(1, 4, 1):
    file_path = path + 'Folder' + str(num) + '/File' + str(num) + '.xlsx'
    data = pd.read_excel(file_path, sheet_name='Sheet1')
    dfs.append(data)
    df = pd.concat(dfs, ignore_index=True)
    df = df.drop_duplicates(subset=['Call Id'], inplace=False)
```

```python
df.dropna(subset=['Remove']) ## Drop any rows missing data in specific column similar to how we have it structed atm ## 
```

```python
df.set_index('Call Id', inplace=True)
```

```python
df.drop(columns=['Remove'], inplace=True)
```
